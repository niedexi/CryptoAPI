```python
import requests

PL = 'https://poloniex.com/public?command=returnOrderBook&currencyPair=all&depth=1'
BF = 'https://api.bitfinex.com/v2/tickers?symbols=ALL'
pol = requests.get(PL).json()
bit = requests.get(BF).json()

```

```python
for p in pol:
    print(pol[p]['asks'])
```

```python
for b in bit:
    print(b[1], b[3])
```
