```python
eth = pd.read_json('https://poloniex.com/public?command=returnChartData&currencyPair=USDT_ETH&start=1503120235&end=1503206635&period=300')
```

```python
btc = pd.read_json('https://poloniex.com/public?command=returnChartData&currencyPair=USDT_BTC&start=1503120235&end=1503206635&period=300')
```

```python
compare = pd.DataFrame({'btc': btc['close'], 'eth': eth['close']})
```

```python
compare.plot(grid=True)
```

```python
compare2 = compare.apply(lambda x: x/x[0])
```

```python
compare2.plot(grid=True).axhline(y=1, color='black', lw=2)
```

```python
compare3 = compare.apply(lambda x: np.log(x) - np.log(x.shift(1)))
compare3.plot(grid = True).axhline(y = 0, color = "black", lw = 2)
```
